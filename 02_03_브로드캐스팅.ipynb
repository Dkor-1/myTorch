```python
import torch
```

```python
a = torch.tensor(
    [
        [1, 2, 3],
        [4, 5, 6],
    ]
)
b = torch.tensor(
    [1, -1, 2],
)

print(f'a의 shape: {a.shape}')
print(f'b의 shape: {b.shape}')
```

```python
a + b
```

```python
b_expanded = torch.tensor(
    [
        [1, -1, 2],
        [1, -1, 2],
    ]
)
a + b_expanded
```

```python
a = torch.tensor(
    [
        [-1, 1],
        [1, 1],
    ]
)
```

```python
a + 1
```

```python
a * 2
```
