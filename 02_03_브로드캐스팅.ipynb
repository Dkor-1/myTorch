```python
import numpy as np
import torch

tensor0 = torch.tensor(
    [
        [1, 1], 
        [-1, -1]
    ]
)
tensor1 = torch.tensor(
    [
        [1, 1], 
        [0, 0]
    ]
)
tensor2 = torch.tensor(
    [
        [1, -1], 
        [1, -1]
    ]
)

# 여기에 코드를 작성하세요.
tensor3 = tensor0 + tensor1
tensor3 = tensor3 @ tensor2
f_tensor = tensor3 + torch.tensor([[2,-2]])
final_np_array = f_tensor.numpy()
# 테스트 코드
print(final_np_array)
```
