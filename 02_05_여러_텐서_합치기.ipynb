```python
import torch
```

```python
x = torch.zeros(2, 3)
y = torch.ones(2, 3)
```

```python
# concatenate
cat0 = torch.cat([x, y], dim=0)
cat1 = torch.cat([x, y], dim=1)
```

```python
print(cat0)
print(cat0.size())
```

```python
print(cat1)
print(cat1.size())
```

```python
x = torch.zeros(2, 3)
y = torch.ones(2, 3)
```

```python
stack_result = torch.stack([x, y], dim=0)
cat_result = torch.cat([x, y], dim=0)
```

```python
print(stack_result.size())
print(stack_result, '\n')

print(cat_result.size())
print(cat_result)
```

```python
stack_result_dim2 = torch.stack([x, y], dim=2)

print(stack_result_dim2.size())
print(stack_result_dim2)
```
